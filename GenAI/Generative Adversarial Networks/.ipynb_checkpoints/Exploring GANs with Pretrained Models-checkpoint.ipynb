```python
#from transformers import BigGAN # ERROR!
import nltk
nltk.download('wordnet')

from pytorch_pretrained_biggan import BigGAN
import torch
from torchvision import transforms
from PIL import Image

# Load pretrained BigGAN model
model = BigGAN.from_pretrained("biggan-deep-256")

# Generate random latent vector (noise)
latent_vector = torch.randn(1, 128)  # 128 is the size of the input vector for BigGAN

# Set class label 
class_vector = torch.zeros(1, 1000)
class_vector[0, 207] = 1

# Generate image
with torch.no_grad():
    generated_image = model(latent_vector, class_vector, truncation=0.5)

# Convert the tensor to a displayable image
generated_image = (generated_image + 1) / 2
generated_image = generated_image.squeeze().permute(1, 2, 0).cpu().numpy()
generated_image = (generated_image * 255).astype('uint8')
Image.fromarray(generated_image)
```

```python
import torch
from pytorch_pretrained_biggan import BigGAN
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# Load BigGAN model
model = BigGAN.from_pretrained('biggan-deep-256')
model.eval()

# Set class (e.g., 207 = golden retriever)
class_vector = torch.zeros(1, 1000)
class_vector[0, 207] = 1

# Store generated images
images = []

# Generate 5 variations of the latent vector
for i in range(5):
    # Generate random latent vector (noise)
    latent_vector = torch.randn(1, 128)

    # Modify a few dimensions differently in each loop
    latent_vector[0, i] = 2    # Boost dimension i

    # Generate image
    with torch.no_grad():
        generated_image = model(latent_vector, class_vector, truncation=0.5)

    # Convert to image format
    generated_image = (generated_image + 1) / 2
    generated_image = generated_image.squeeze().permute(1, 2, 0).cpu().numpy()
    img = Image.fromarray((generated_image * 255).astype(np.uint8))

    # Save image
    img.save(f"generated_image_{i+1}.png")
    
    images.append(img)

# Display all 5 images in a row
plt.figure(figsize=(15, 3))
for i, img in enumerate(images):
    plt.subplot(1, 5, i + 1)
    plt.imshow(img)
    plt.axis('off')
    plt.title(f'Image {i+1}')
plt.show()
```

In this project, I explored how modifying the latent vector (the 128-dimensional input noise) affects image generation using the BigGAN model. Instead of traditional image generation methods that rely on structured templates or algorithms, BigGAN generates images entirely from noise and class input, using a pretrained model trained on ImageNet. The goal was to understand how changing values within the latent vector impacts the diversity, realism, and interpretability of the generated outputs.

To conduct this experiment, I used the pytorch_pretrained_biggan library to load the BigGAN-deep-256 model. It’s important to note that BigGAN is not available in the Hugging Face transformers library. Attempting from transformers import BigGAN results in an import error because transformers focuses on models like BERT, GPT, and CLIP, not vision-only GANs like BigGAN. Instead, I correctly used from pytorch_pretrained_biggan import BigGAN, which provides pretrained BigGAN models and utility functions.

For image generation, I fixed the class vector to represent “golden retriever” (ImageNet class index 207) and created five different latent vectors by sampling from a normal distribution. In each iteration, I modified one dimension of the latent vector (e.g., setting latent_vector[0, i] = 2) to explore how subtle changes in the input influence the output. All other parameters, including truncation (set to 0.5), were kept constant to isolate the effect of latent manipulation.

The generated images showed noticeable differences in structure, color, and clarity, even though the class label remained the same. Some images resembled golden retrievers quite well, while others included distorted shapes, cluttered backgrounds, or less recognizable content. This inconsistency reflects how sensitive BigGAN is to both the latent vector and truncation values. It also suggests that the latent space encodes complex and abstract visual feature - modifying just a few dimensions can affect pose, texture, or object composition.

One key observation was that higher or extreme values in certain latent dimensions often disrupted image quality, introducing artifacts or noise. On the other hand, when changes were subtle or within a reasonable range, the images maintained more recognizable and realistic forms. However, even within this range, some outputs looked unusual or dreamlike, reinforcing the idea that GANs do not guarantee consistent realism.

Compared to traditional methods, this approach to image generation is both more flexible and less predictable. There is no explicit control over features like eye color, background, or lighting. Instead, BigGAN interprets the latent vector as a high-level representation, mapping it to learned image features from training. This makes it a powerful tool for generating variety, but also highlights challenges in interpretability and reliability.

Overall, this project showed that GANs like BigGAN are capable of producing diverse and often realistic images from noise, but their outputs are heavily influenced by both the structure of the latent space and the class conditioning. While powerful, this method lacks precision and requires experimentation to achieve desired results. The experience also emphasized the importance of using the correct library (pytorch_pretrained_biggan) and understanding the model's expectations for input formatting to avoid errors and unexpected results.
